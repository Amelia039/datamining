**Rumus Mean**

Dalam statistika dan matematika, mean disebut juga dengan rata-rata. Jadi, definisi mean adalah nilai rata-rata dari sekumpulan data yang terdiri dari dua atau lebih..
Berikut rumus mean untuk data tunggal:
$$\bar{x}=\frac{\sum x_i}{n}
$$
Keterangan:

$\bar{x}:$ nilai rata-rata

$\mathrm{x}_{\mathrm{i}}$ : nilai data ke-i

n : banyaknya data

```python
mean = {'Pregnancies':np.mean(c['Pregnancies']),
		'Glucose': np.mean(c['Glucose']),'BloodPressure':np.mean(c['BloodPressure']),
    'SkinThickness':np.mean(c['SkinThickness']),'Insulin':np.mean(c['Insulin']),
    'BMI':np.mean(c['BMI']),'DiabetesPedigreeFunction':np.mean(c['DiabetesPedigreeFunction']),
    'Age':np.mean(c['Age']),'Outcome':np.mean(c['Outcome'])
    }

df = pd.DataFrame(mean, index=['mean'])
df
```

**Rumus Median**

Median disebut juga dengan nilai tengah. Median diperoleh dengan cara mengambil nilai tengah pada data yang telah disusun secara berurutan dari angka terkecil ke terbesar.
Pada data tunggal untuk menentukan median yaitu:
Misalkan banyaknya data n, 

Untuk $n$ ganjil: $M e=x_{\frac{1}{2}(n+1)}$

Untuk $n$ genap: $M e=\frac{\frac{x_n}{2}+\frac{x_n}{2}+1}{2}$

Keterangan: 
$x_{\frac{n}{2}}=$ data pada urutan ke- $\frac{n}{2}$ setelah diurutkan.

```python
median = {'Pregnancies':np.median(c['Pregnancies']),
		'Glucose': np.median(c['Glucose']),'BloodPressure':np.median(c['BloodPressure']),
    'SkinThickness':np.median(c['SkinThickness']),'Insulin':np.median(c['Insulin']),
    'BMI':np.median(c['BMI']),'DiabetesPedigreeFunction':np.median(c['DiabetesPedigreeFunction']),
    'Age':np.median(c['Age']),'Outcome':np.median(c['Outcome'])
    }

df = pd.DataFrame(median, index=['median'])
df
```

**Modus**

Modus adalah nilai data yang paling sering muncul atau nilai data yang frekuensinya paling besar.

```python
modus = {'Pregnancies':stats.mode(c['Pregnancies']),
		'Glucose': stats.mode(c['Glucose']),'BloodPressure':stats.mode(c['BloodPressure']),
    'SkinThickness':stats.mode(c['SkinThickness']),'Insulin':stats.mode(c['Insulin']),
    'BMI':stats.mode(c['BMI']),'DiabetesPedigreeFunction':stats.mode(c['DiabetesPedigreeFunction']),
    'Age':stats.mode(c['Age']),'Outcome':stats.mode(c['Outcome'])
    }

modus
```

**Rumus Standar Deviasi**

Standar deviasi atau simpangan baku biasanya diajarkan pada ilmu statistik untuk mengukur tingkat kesamaan atau kedekatan dalam suatu kelompok.

Standar deviasi adalah nilai statistik yang sering kali dipakai dalam menentukan kedekatan sebaran data yang ada di dalam sampel dan seberapa dekat titik data individu dengan mean atau rata-rata nilai dari sampel itu sendiri.

Untuk Data Tunggal :
$S=\sqrt{\frac{\sum\left(x_i-\bar{x}\right)^2}{n}}$

$S=$ Standar deviasi

$x_i=$ nilai $\mathrm{x} \mathrm {ke-i}$

$\bar{x}=$ nilai rata-rata data

$n=$ jumlah data

```python
std = {'Pregnancies':st.stdev(c['Pregnancies']),
		'Glucose':st.stdev(c['Glucose']),'BloodPressure':st.stdev(c['BloodPressure']),
    'SkinThickness':st.stdev(c['SkinThickness']),'Insulin':st.stdev(c['Insulin']),
    'BMI':st.stdev(c['BMI']),'DiabetesPedigreeFunction':st.stdev(c['DiabetesPedigreeFunction']),
    'Age':st.stdev(c['Age']),'Outcome':st.stdev(c['Outcome'])
    }

df = pd.DataFrame(std, index=['std'])
df
```

RUMUS PERSENTIL DATA TUNGGAL
$$
P_i=\text { data } k e-\frac{i(n+1)}{100}
$$

Jika persentil 25% maka:
$$
P_i=\text { data } k e-\frac{25(n+1)}{100}
$$

Jika persentil 50% maka:
$$
P_i=\text { data } k e-\frac{50(n+1)}{100}
$$

Jika persentil 75% maka:
$$
P_i=\text { data } k e-\frac{75(n+1)}{100}
$$

Keterangan:

$\mathrm{P}=$ persentil ke $-\mathrm{i}$

$$
i=1,2,3, \ldots, 99
$$
n = banyak data



```python
percentile25 = {'Pregnancies':np.percentile(c['Pregnancies'],25),
		'Glucose':np.percentile(c['Glucose'],25),'BloodPressure':np.percentile(c['BloodPressure'],25),
    'SkinThickness':np.percentile(c['SkinThickness'],25),'Insulin':np.percentile(c['Insulin'],25),
    'BMI':np.percentile(c['BMI'],25),'DiabetesPedigreeFunction':np.percentile(c['DiabetesPedigreeFunction'],25),
    'Age':np.percentile(c['Age'],25),'Outcome':np.percentile(c['Outcome'],25)
    }

df = pd.DataFrame(percentile25, index=['25%'])
df
```
