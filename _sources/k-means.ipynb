```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
#the imported dataset does not have the required column names so lets add it
colnames = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
irisdata = pd.read_csv(url, names=colnames)
irisdata
```

```python
kelas=irisdata['Class'] = pd.Categorical(irisdata["Class"])
data=irisdata["Class"] = irisdata["Class"].cat.codes
data
```

```python
kelas
```

```python
X = irisdata.values[:, 0:4]
y = irisdata.values[:, 4]
```

```python
from sklearn.cluster import KMeans

# Number of clusters
kmeans = KMeans(n_clusters=3)
# Fitting the input data
kmeans = kmeans.fit(X)

# Getting the cluster labels
labels = kmeans.predict(X)
# Centroid values
centroids = kmeans.cluster_centers_
```

```python

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(labels)
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_name_mapping)
{'1': 0, '0': 1, '2': 2}
```

```python
databaru=np.array(le_name_mapping)
databaru
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(labels,y)
```
